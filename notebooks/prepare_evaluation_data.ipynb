```python
from prompts import general_company_info, assistant_identity_context
prompt = assistant_identity_context + general_company_info
```

```python
question_dicts = [
    {"messages": [
        {
        "role": "system",
        "content": prompt
        },
        {
            "role": "user",
            "content": q.question
        },
        {
            "role": "assistant",
            "content": q.response
        }
    ] }for q in prepared_questions]

question_dicts
```

```python
output_file_path = '../data/validation_set.jsonl'
with open(output_file_path, 'w') as f:
    for question_dict in question_dicts:
        json.dump(question_dict, f)
        f.write('\n')
```
